```python
import matplotlib.pyplot as plt
def collatz(n):
    step =[n]
    while n != 1:
        if (n % 2 == 0):
            n /= 2
            print(n)
        else:
            n = 3 * n + 1
            print(n)
        step.append(n)
    return (step)
x=[]
y=[]
for i in range(1,1000):
    step = collatz(i)
    x.append(i)
    y.append(len(step)-1) 
plt.plot(x,y,"o")
plt.show()
```
